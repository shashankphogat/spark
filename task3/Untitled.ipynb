```python
import pandas as pd
```

```python
data=pd.read_csv("SampleSuperstore.csv")
```

```python
data.head()
```

```python
data=data.dropna()
data=data.drop(['Country','Postal Code'],axis=1)
data.head()
```

```python
data2=data.groupby(['Region']).sum()
data2.sort_values('Profit',ascending=False)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig=plt.figure(figsize=(18,10))
fig.suptitle('Region Wise')
plt1=fig.add_subplot(2,2,1)
plt1.title.set_text('Profit')
plt1.bar(data2.index, data2.Profit,0.5)
plt2=fig.add_subplot(2,2,2)
plt2.title.set_text('Discount')
plt2.bar(data2.index, data2.Discount,0.5,color='orange')
plt3=fig.add_subplot(2,2,(3,4))
plt3.title.set_text('Sales')
plt3.bar(data2.index, data2.Sales,0.5,color='green')
plt.show()
```

```python
data3=data.groupby(['Category']).sum()
data3.sort_values('Profit',ascending=False)
```

```python
fig=plt.figure(figsize=(18,10))
fig.suptitle('Category Wise')
plt1=fig.add_subplot(2,2,1)
plt1.title.set_text('Profit')
plt1.bar(data3.index, data3.Profit,0.5)
plt2=fig.add_subplot(2,2,2)
plt2.title.set_text('Discount')
plt2.bar(data3.index, data3.Discount,0.5,color='orange')
plt3=fig.add_subplot(2,2,(3,4))
plt3.title.set_text('Sales')
plt3.bar(data3.index, data3.Sales,0.5,color='green')
plt.show()
```
